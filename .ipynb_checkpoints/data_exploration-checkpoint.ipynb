```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%config InlineBackend.figure_format = 'retina'
plt.style.use(['seaborn-v0_8-colorblind', 'seaborn-v0_8-darkgrid'])
```

```python
baby_df = pd.read_csv('data/Baby_Huckleberry_Data--2026-02-16.csv')
```

```python
baby_df.info()
```

```python
baby_df.head()
```

```python
baby_df['Start'] = pd.to_datetime(baby_df['Start'])
baby_df['End'] = pd.to_datetime(baby_df['End'])

is_sleep = baby_df['Type'] == 'Sleep'
baby_df.loc[is_sleep, 'Sleep_Duration'] = pd.to_timedelta(baby_df.loc[is_sleep, 'Duration'] + ':00').dt.total_seconds() / 60

is_feed = baby_df['Type'] == 'Feed'
baby_df.loc[is_feed, 'Feed_mL'] = pd.to_numeric(baby_df.loc[is_feed, 'End Condition'].str.extract(r'(\d+\.?\d*)')[0])
is_oz = is_feed & baby_df['End Condition'].str.contains('oz', case=False)
baby_df.loc[is_oz, 'Feed_mL'] *= 29.57
```

```python
plt.scatter(baby_df['Start'], baby_df['Feed_mL'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
baby_df['Feed_mL'] = baby_df['Feed_mL'].replace(0, np.nan)
baby_df['Feed_mL_norm'] = (baby_df['Feed_mL'] - baby_df['Feed_mL'].mean()) / baby_df['Feed_mL'].std()
baby_df['Days'] = (baby_df['Start'] - baby_df['Start'].min()).dt.days
baby_df['Days_norm'] = (baby_df['Days'] - baby_df['Days'].mean()) / baby_df['Days'].std()
```

```python
plt.scatter(baby_df['Days_norm'], baby_df['Feed_mL_norm'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
import pymc as pm

baby_feed_df = baby_df.dropna(subset=['Days_norm', 'Feed_mL_norm'])

with pm.Model() as eating_model:
    days = pm.Data('days', baby_feed_df['Days_norm'].values)
    feed_obs = pm.Data('feed_obs', baby_feed_df['Feed_mL_norm'].values)
    
    a = pm.Normal('a', mu=0, sigma=1)
    b = pm.Normal('b', mu=1, sigma=1)
    
    sigma = pm.Exponential('sigma', 1)
    mu = pm.Deterministic('mu', a + b*days)
    feed = pm.Normal('feed', mu=mu, sigma=sigma, observed=feed_obs)

    idata_eating = pm.sample(1000, tune=1000)
    idata_eating.extend(pm.sample_posterior_predictive(idata_eating))
    
```

```python
import arviz as az

feed_mean = baby_feed_df['Feed_mL'].mean()
feed_std = baby_feed_df['Feed_mL'].std()
idata_eating.posterior['y_model'] = (idata_eating.posterior['a'] + idata_eating.posterior['b']*idata_eating.constant_data['days']) * feed_std + feed_mean
az.plot_lm(idata=idata_eating, x=baby_feed_df['Days'], y=baby_feed_df['Feed_mL'], y_model='y_model', kind_pp='hdi', kind_model='hdi');
```

```python
az.summary(idata_eating, var_names=['a', 'b'])
```

```python
sleep_time_df = baby_df[baby_df['Type'] == 'Sleep']
sleep_time_df['Nap_Start'] = (sleep_time_df['Start'] - sleep_time_df['Start'].min()).dt.total_seconds() / 60
sleep_time_df['Internap_Times'] = sleep_time_df['Nap_Start'].sort_values(ascending=True).diff()
sleep_time_df = sleep_time_df.dropna(subset=['Internap_Times'])
sleep_time_df.head()
```

```python
sleep_time_df = sleep_time_df[sleep_time_df['Internap_Times'] < 240]
sleep_time_df['Internap_Times'] = sleep_time_df['Internap_Times'].astype(int)
```

```python
with pm.Model() as poisson_sleep:
    sleep_times = pm.Data('sleep_times', sleep_time_df['Internap_Times'])
    
    sigma = pm.TruncatedNormal('sigma', mu=30, sigma=5, lower=0)
    mu = pm.Normal('mu', mu=150, sigma=30)
    
    nap_lags = pm.StudentT('nap_lags', mu=mu, sigma=sigma, nu=sleep_times.shape[0]-1, observed=sleep_times)

    idata = pm.sample_prior_predictive(draws=100)
    idata.extend(pm.sample(1000, tune=1000, idata_kwargs={'log_likelihood': True}))
    idata.extend(pm.sample_posterior_predictive(idata))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot Prior Predictive
az.plot_ppc(idata, group='prior', kind='kde', observed=True, ax=ax1, color='lightgray');
ax1.set_title('Prior Predictive (Pre-Data)');

# Plot Posterior Predictive
az.plot_ppc(idata, group='posterior', kind='kde', observed=True, ax=ax2, color='lightgray');
ax2.set_title('Posterior Predictive (Post-Data)');
```
